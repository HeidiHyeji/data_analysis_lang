```python
#파이썬의 연산자 사용
#변수와 상수 파이썬의 다양한 연산자를 이용하여 원의 면적 구하기,
#두 사각형의 크기 비교, 성적 구하기 와 같은 산수 문제
#1. 반지름이 5.5인 원의 면적을 계산(원주율은 3.14)

PI = 3.14
r = 5.5
area = r*r*PI
print(area);
```

```python
#3. 철수는 중간고사에서 국어 95점, 수학 88점, 영어 90점을 받았고,
#   과학 점수는 잊어버렸지만 네 과목 평균 점수는 92점, 철수의 과학점수는?
korean = 95
math = 88
english = 90

science = (92 * 4) - (korean + math + english)
print(science)
```

```python
# 음료자판기
# 0이거나 -1 입력시 종료

#orange = 100
#coffe = 200
#coke = 300
drink = {'오렌지주스':100,'커피':200,'콜라': 300}

while True:
  print('음료목록 1.오렌지주스(100원), 2.커피(200원), 3.콜라"(300원)')
  coin = int(input('동전을 넣으세요.(종료 시, 0 또는 -1 입력)'))
  if coin == 0 or coin == -1:
    break
  order = int(input('음료를 고르세요.\n'))

  if order < 1 or order > 3 :
    print("없는 메뉴입니다. 다시 입력해 주세요.\n")
    continue
  elif coin < list(drink.values())[order-1] :
    print("잔액이 부족합니다.\n")
  else:
    print('{}가 곧 제공됩니다.'.format(list(drink.keys())[order-1]))
    print("거스름돈은 {}원입니다.\n".format(coin-list(drink.values())[order-1]))
```
