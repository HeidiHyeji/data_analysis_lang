# [LAB] 소수 판별기
- 사용자에게 숫자를 입력 받고 숫자가 소수(Prime number)인지 판별하는 프로그램

```python
def check_prime_num(x): # 소수 판별
    for i in range(2, x):
        if x % i == 0:
            # x가 i로 나누어 떨어지면 실행하기
            return False;
        return True;

while True:
    number = int(input('판별한 자연수를 입력하세요:'))
    if number == -1:
        print('종료')
        break;
    print(check_prime_num(number))
```

# [실전 분석] 예금 이자 계산기

```python
def interest_year(p,r,n):
    return p * (1+r) ** n # 원금 p, 연간 이율 r, 거치 연수 n을 입력 받아 원리금 result를 반환

p = 30000000
r = 0.051
n = 3

result = interest_year(p,r,n)
print('원금: {0}, 이자: {1}'.format(p, result-p))
```
